```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

##1. Complex Algebra 1 - Left and right-hand side equations

```python
def left_equ(y):
  x = np.linspace(-5,5,5)
  z = x + y*1j
  ans = np.conj(z)
  plt.plot(ans)
  return ans

def right_equ(y):
  x = np.linspace(-5,5,5)
  z = x + y*1j
  ans = 1/z
  plt.plot(z)
  return ans
```

```python
left_equ(5) == right_equ(5)
```

```python
left_equ(5)
```

##2. Complex Algebra 2 - Original and simplified versions of x[n]

```python
def plot_simp(theta):
  x = np.linspace(-5,5, 50)
  xn = ((math.e ** (math.pi/2 * 1j) + math.e ** (math.pi/2 * -1j))/2) - ((math.e**1j*x*theta) - (math.e**-1j*x*theta))/1j
  xnsimp = 1j*((math.e**1j*theta*x)-(math.e**-1j*theta*x))
  plt.plot(xn, color="red", label="xn")
  plt.plot(xnsimp, color="blue", label="simplified")
  plt.legend(loc="upper left")
  return xn
```

```python
plot_simp(90)
```

##3.  Periodic Signals

```python
def plot_period(omega):
  x = np.linspace(-5,5,100)
  y = np.exp(omega*1j*2*np.pi*x)
  plt.plot(x,y)
  plt.show()
```

```python
a, b, c = 1, 1/3, 56/9
```

```python
plot_period(a)
```

```python
plot_period(b)
```

```python
plot_period(c)
```

##4. Signal Operations 1 - Moving average

```python
def mov_ave(list):
  i = 0
  moving_average = []
  while i < len(list)-1:
    average = round(np.sum(list[i:i+2]))/2
    moving_average.append(average)
    i +=1
  plt.title("Moving Average")
  plt.plot(list, color="blue", label="xn")
  plt.plot(moving_average, color="red", label="yn")
  plt.legend(loc="upper left")
  return moving_average
```

```python
list = [0,1,2,3,0]
mov_ave(list)
```

##5. Signal Operations - Input-output relationship

```python
def outres(list, a, b, c):
  output = []
  i = 0
  while i < len(list):
    yn = b * (a + list[i]-1) - (list[i]-3*(c*list[i] + list[i]))
    output.append(yn)
    i+=1
  plt.plot(list, color="blue", label="xn")
  plt.plot(output, color="red", label="yn")
  plt.legend(loc="upper left")
  return output
```

```python
list = [1,2,3]
outres(list, 2, 3, 5)
```
